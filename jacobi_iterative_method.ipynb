```python
import pandas as np
import numpy as np
import timeit
```

```python
def jacobi_iterative_method(A, B, tolerance, max_iters=100):

    #check if diagonal dominance
    if not np.all(np.diagonal(A) > np.sum(np.abs(A), axis=1) - np.diagonal(A)):
        raise ValueError("Matrix is not diagonal dominant")
    
    n = A.shape[0]
    X = np.zeros((max_iters, B.shape[0]))
    X[:, 0] = 0
    k = 0

    while True:
        if k > max_iters:
            break
        k+=1
        for i in range(n):
            X[k, i] = (B[i] - (np.dot(A[i], X[k-1]) - A[i, i] * X[k-1, i]))/A[i, i]
        if np.all(np.abs(X[k] - X[k-1]) < tolerance):
            break
    
    return X, k

A = np.array([[9, 2, 3], [1, 12, 9], [4, 6, 14]])
B = np.array([7, 2, 1]).reshape(-1, 1)
tolerance = 1e-6
max_iters = 100

X, k = jacobi_iterative_method(A, B, tolerance, max_iters)
```
