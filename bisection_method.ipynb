```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: -np.sin(x**10)**2 - 0.9*x**2 + 7*x - 4 - np.tan(4*x**10)**2 + (1/np.sin(4*x**2)**2)
f_x = [f(x) for x in np.arange(-100, 100, 1)]

plt.plot(np.arange(-100, 100, 1), f_x)
```

```python

def bisection_method(f, a, b, tol=1e-9, max_iters=100):
    fa, fb = f(a), f(b)
    if fa * fb > 0:
        raise ValueError("f(a) and f(b) must have opposite signs")
    
    for i in range(1, max_iters+1):
        c   = 0.5*(a + b)
        fc  = f(c)
        # 2) check for convergence
        if abs(fc) < tol or (b - a)/2 < tol:
            return c, i
        
        if fa * fc < 0:
            b, fb = c, fc
        else:
            a, fa = c, fc
    
    return c, max_iters

bisection_method(f, 0, 9)
```
