```python
f = lambda x: -np.sin(x**10)**2 - 0.9*x**2 + 7*x - 4 - np.tan(4*x**10)**2 + (1/np.sin(4*x**2)**2)


def gradient_of_f(x0, x1):
    return (f(x0) - f(x1)) / (x0-x1)

def secant_method(f, x0, x1, tol=1e-4, max_iters=100):
    k = 1
    while k <= max_iters:
        try:
            f_prime = gradient_of_f(x0, x1)
            if abs(f_prime) < 1e-12:
                raise ZeroDivisionError("Derivative near zero; method may diverge.")
            x2 = x1 - f(x1)/f_prime
        except Exception as e:
            print(f"Error during iteration {k}: {e}")
            return None, k

        if abs(x2 - x1) < tol:
            return x2, k

        x0, x1 = x1, x2
        k += 1
    return x2, k

secant_method(f, 2, 5)
```
